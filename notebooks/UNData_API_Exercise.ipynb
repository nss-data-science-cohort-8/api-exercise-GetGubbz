In this exercise, you'll redo the data gathering phase of the UNData Exploration project by using APIs instead of downloading csv files.

1. Use the API to get all available data for the _GDP per capita, PPP (constant 2017 international $)_ indicator. Hint: this indicator has code "NY.GDP.PCAP.PP.KD". Adjust the query parameters so that you can retrieve all available rows. Convert the results to a DataFrame.

```python
import requests
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
endpoint_GDP_PPP = 'https://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD?format=json'
params = {
    'per_page':20000
}
response_GDP = requests.get(endpoint_GDP_PPP, params=params)
```

```python
response_GDP
```

```python
res_GDP = response_GDP.json()
res_GDP
```

```python
res_GDP_df = pd.json_normalize(res_GDP[1])
res_GDP_df = pd.DataFrame(data = res_GDP_df)
```

```python
res_GDP_df.shape
```

2. Now, use the API to get all available data for _Life expectancy at birth, total (years)_. This indicator has code "SP.DYN.LE00.IN". Again, convert the results to a DataFrame.

```python
endpoint_le = 'https://api.worldbank.org/v2/country/all/indicator/SP.DYN.LE00.IN?format=json'
params = {
    'per_page':20000
}
response_le = requests.get(endpoint_le, params=params)
```

```python
response_le
```

```python
res_le = response_le.json()
res_le
```

```python
res_le_df = pd.json_normalize(res_le[1])
res_le_df = pd.DataFrame(data = res_le_df)
```

```python
res_le_df.head()
```

3. Merge the two results DataFrames together. You may want to rename or drop columns prior to merging.

```python
res_le_df = res_le_df.drop(columns = ['countryiso3code', 'unit', 'indicator.id', 'decimal', 'obs_status', 'indicator.value'], axis = 1)
res_le_df = res_le_df.rename(columns = {'value':'Life_Expectancy', 'country.value':'Country', 'date':'Year', 'country.id':'Country_ID'})
res_GDP_df = res_GDP_df.drop(columns = ['countryiso3code', 'unit', 'indicator.id', 'decimal', 'obs_status', 'indicator.value'], axis = 1)
res_GDP_df = res_GDP_df.rename(columns = {'value':'GDP_Per_Capita', 'country.value':'Country', 'date':'Year', 'country.id':'Country_ID'})
```

```python
res_le_df.head()
```

```python
res_GDP_df.head()
```

```python
gdp_le = pd.merge(res_le_df, res_GDP_df, how = 'inner', on = ['Year', 'Country_ID', 'Country'])
```

```python
gdp_le.shape
```

4. You can also get more information about the available countries (region, capital city, income level classification, etc.) by using the [Country API](https://datahelpdesk.worldbank.org/knowledgebase/articles/898590-country-api-queries). Use this API to pull in all available data. Merge this with your other datasets. Use this to now remove the rows that correspond to regions and not countries.

```python
endpoint_country = 'https://api.worldbank.org/v2/country?format=json'
params = {
    'per_page':100000
}
response_country = requests.get(endpoint_country, params=params)
```

```python
res_country = response_country.json()
res_country
```

```python
res_country_df = pd.json_normalize(res_country[1])
res_country_df = pd.DataFrame(data = res_country_df).rename(columns = {'name':'Country'})
```

```python
res_country_df.head()
```

```python
res_country_df.shape
```

```python
gdp_le_countries = pd.merge(gdp_le, res_country_df, how = 'inner', on = 'Country', indicator=True, validate = 'many_to_many')
gdp_le_countries.head()
```

```python
gdp_le_countries['capitalCity'].value_counts()
```

```python
gdp_le_countries['capitalCity'] == ''
```

```python
gdp_le_countries = gdp_le_countries[gdp_le_countries['capitalCity'] != '']
gdp_le_countries.head()
```

```python
gdp_le_countries[['region.id','region.value']].value_counts()
```
